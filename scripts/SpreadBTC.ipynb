```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

def get_btc_data(vs_currency='usd', days=90):
    url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart'
    params = {'vs_currency': vs_currency, 'days': days}
    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()
    prices = data['prices']
    df = pd.DataFrame(prices, columns=['timestamp', f'price_{vs_currency}'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    return df

btc_usd = get_btc_data('usd')
btc_eur = get_btc_data('eur')

btc_data = pd.merge(btc_usd, btc_eur, on='timestamp')

btc_data['spread'] = btc_data['price_usd'] - btc_data['price_eur']
btc_data['day'] = btc_data['timestamp'].dt.day
btc_data['hour'] = btc_data['timestamp'].dt.hour

btc_data.to_csv('../data/btc_spread.csv', index=False)
print("âœ… Archivo guardado: ../data/btc_spread.csv")

pivot = btc_data.pivot_table(values='spread', index='hour', columns='day', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot, cmap="Blues", center=0)
plt.title('Spread BTC/USD - BTC/EUR por hora y dÃ­a')
plt.xlabel('DÃ­a del mes')
plt.ylabel('Hora del dÃ­a')
plt.show()

hourly_spread = btc_data.groupby('hour')['spread'].mean()

max_hour = hourly_spread.idxmax()
min_hour = hourly_spread.idxmin()

print(f"Hora con mayor spread promedio: {max_hour}:00")
print(f"Hora con menor spread promedio: {min_hour}:00")


```
