```python
import numpy as np
import pandas as pd

df = pd.read_csv("../../data/btc_exchange_rates.csv")

df["date"] = pd.to_datetime(df["date"])

df
```

```python
price_usd = df["price_usd"]
price_eur = df["price_eur"]
```

```python
usd_initial = price_usd.iloc[0]
usd_final = price_usd.iloc[-1]

eur_initial = price_eur.iloc[0]
eur_final = price_eur.iloc[-1]
```

```python
usd_return = (usd_final - usd_initial) / usd_initial * 100
eur_return = (eur_final - eur_initial) / eur_initial * 100

print(f"Rendimiento acumulado BTC/USD: {usd_return:.2f}%")
print(f"Rendimiento acumulado BTC/EUR: {eur_return:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df["date"], df["price_usd"], label="BTC/USD", color="blue")
plt.plot(df["date"], df["price_eur"], label="BTC/EUR", color="orange")

plt.title("Evolución del precio de BTC en USD y EUR")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()
plt.grid(True)
plt.show()
```
